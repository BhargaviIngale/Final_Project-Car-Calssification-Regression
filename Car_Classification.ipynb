```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('cars_class.csv')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
x = data.drop(['ID','Class'],axis=1)
y = data['Class']
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Scaling the training data using standard scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
```

```python
#Scaling test data using standard scaler
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier

LR=LogisticRegression()
KNN=KNeighborsClassifier()
svc=SVC()
DTC=DecisionTreeClassifier()
ABC=AdaBoostClassifier()
GBC=GradientBoostingClassifier()
RFC=RandomForestClassifier()
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
#makind an evaluation function
def evaluate_preds(y_true,y_preds):
    accuracy = accuracy_score(y_test,y_preds)

    f1_weighted = f1_score(y_test,y_preds,average='weighted')
    f1_macro = f1_score(y_test,y_preds,average='macro')
    f1_micro = f1_score(y_test,y_preds,average='micro')
    confusion_matrix_result = confusion_matrix(y_true,y_preds)
    metric_dict = {'accuracy':round(accuracy,2),

                   'f1_weighted':round(f1_weighted,2),
                   'f1_macro':round(f1_macro,2),
                   'f1_micro':round(f1_micro,2),
                   'confusion_matrix':confusion_matrix_result}
    print(f'Acc: {accuracy*100:.2f}%')

    print(f'F1_weighted: {f1_weighted:.2f}')
    print(f'F1_macro: {f1_macro:.2f}')
    print(f'F1_micro: {f1_micro:.2f}')
    print(f'Confusion Matrix: \n {confusion_matrix_result}')

    return metric_dict
```

```python
np.random.seed(42)
#fitting the model
LR.fit(x_train,y_train)
#making predictions
y_preds = LR.predict(x_test)
#Evaluating on the test set
evaluate_preds(y_test,y_preds);
```

```python
np.random.seed(42)
#fitting the model
KNN.fit(x_train,y_train)
#making predictions
y_preds = KNN.predict(x_test)
#Evaluating on the test set
evaluate_preds(y_test,y_preds);
```

```python
np.random.seed(42)
#fitting the model
svc.fit(x_train,y_train)
#making predictions
y_preds = svc.predict(x_test)
#Evaluating on the test set
evaluate_preds(y_test,y_preds);
```

```python
np.random.seed(42)
#fitting the model
DTC.fit(x_train,y_train)
#making predictions
y_preds = DTC.predict(x_test)
#Evaluating on the test set
evaluate_preds(y_test,y_preds);
```

```python
np.random.seed(42)
#fitting the model
RFC.fit(x_train,y_train)
#making predictions
y_preds = RFC.predict(x_test)
#Evaluating on the test set
evaluate_preds(y_test,y_preds);
```

```python
np.random.seed(42)
#fitting the model
GBC.fit(x_train,y_train)
#making predictions
y_preds = GBC.predict(x_test)
#Evaluating on the test set
evaluate_preds(y_test,y_preds);
```
